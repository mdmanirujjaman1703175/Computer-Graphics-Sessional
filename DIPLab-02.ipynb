```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
img_gray=cv2.imread('image.png',0)
img_gray.shape
print(img_gray)
img_gray2=np.uint8(img_gray/128)
print(img_gray2)
img_gray3=img_gray2*255
print(img_gray3)
plt.imshow(img_gray3, cmap='gray')
```

```python
# Function to reduce bit depth
def reduce_bit_depth(img, bits):
    levels = 2 ** bits
    step = 256 // levels

    # new image  (same size)
    reduced_img = np.zeros_like(img)

    rows, cols = img.shape

    for i in range(rows):
        for j in range(cols):
            reduced_img[i, j] = (img[i, j] // step) * step

    return reduced_img
# Apply different bit reductions
img_8bit = img_gray                      # Original (8 bits, 256 levels)
img_7bit = reduce_bit_depth(img_gray, 7) 
img_5bit = reduce_bit_depth(img_gray, 5) 
img_3bit = reduce_bit_depth(img_gray, 3) 
img_2bit = reduce_bit_depth(img_gray, 2)
img_1bit = reduce_bit_depth(img_gray, 1) 

# Show results
plt.figure(figsize=(14,8))

plt.subplot(2,3,1)
plt.imshow(img_8bit, cmap="gray")
plt.title("Original 8-bits (256 levels)")
plt.axis("off")

plt.subplot(2,3,2)
plt.imshow(img_7bit, cmap="gray")
plt.title("7-bits (128 levels)")
plt.axis("off")

plt.subplot(2,3,3)
plt.imshow(img_5bit, cmap="gray")
plt.title("6-bits (32 levels)")
plt.axis("off")

plt.subplot(2,3,4)
plt.imshow(img_3bit, cmap="gray")
plt.title("4-bits (8 levels)")
plt.axis("off")

plt.subplot(2,3,5)
plt.imshow(img_2bit, cmap="gray")
plt.title("2-bits (4 levels)")
plt.axis("off")

plt.subplot(2,3,6)
plt.imshow(img_1bit, cmap="gray")
plt.title("1-bit (2 levels)")
plt.axis("off")

plt.tight_layout()
plt.show()
```
